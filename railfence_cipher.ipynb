```python
def encrypt_rail_fence(text, rails):
    if rails <= 1:
        return text

    fence = [['' for _ in range(len(text))] for _ in range(rails)]
    rail = 0
    direction = 1  # 1: down, -1: up

    for i, char in enumerate(text):
        fence[rail][i] = char
        rail += direction

        if rail == 0 or rail == rails - 1:
            direction *= -1

    encrypted_text = ''.join([''.join(row) for row in fence])
    return encrypted_text.replace('', '')

def decrypt_rail_fence(cipher, rails):
    if rails <= 1:
        return cipher

    fence = [['' for _ in range(len(cipher))] for _ in range(rails)]
    rail = 0
    direction = 1

    for i in range(len(cipher)):
        fence[rail][i] = '*'
        rail += direction

        if rail == 0 or rail == rails - 1:
            direction *= -1

    index = 0
    for row in range(rails):
        for col in range(len(cipher)):
            if fence[row][col] == '*' and index < len(cipher):
                fence[row][col] = cipher[index]
                index += 1

    rail = 0
    direction = 1
    decrypted_text = []

    for i in range(len(cipher)):
        decrypted_text.append(fence[rail][i])
        rail += direction

        if rail == 0 or rail == rails - 1:
            direction *= -1

    return ''.join(decrypted_text)

message = "HELLOPYTHON"
rails = 3

encrypted = encrypt_rail_fence(message, rails)
print(f"Encrypted: {encrypted}")

decrypted = decrypt_rail_fence(encrypted, rails)
print(f"Decrypted: {decrypted}")
```
