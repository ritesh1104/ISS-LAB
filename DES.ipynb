```python
!pip install pycryptodome
from Crypto.Cipher import DES
import binascii

def pad(text):
    """Pads text to be a multiple of 8 bytes for DES encryption."""
    while len(text) % 8 != 0:
        text += ' '
    return text

def des_encrypt(plain_text, key):
    """Encrypts the given plain text using DES."""
    cipher = DES.new(key, DES.MODE_ECB)
    padded_text = pad(plain_text)
    encrypted_text = cipher.encrypt(padded_text.encode())
    return binascii.hexlify(encrypted_text).decode()

def des_decrypt(encrypted_text, key):
    """Decrypts the given encrypted text using DES."""
    cipher = DES.new(key, DES.MODE_ECB)
    decrypted_text = cipher.decrypt(binascii.unhexlify(encrypted_text))
    return decrypted_text.decode().strip()

key = b'8charKey'
plaintext = "HelloColab"

encrypted_text = des_encrypt(plaintext, key)
decrypted_text = des_decrypt(encrypted_text, key)

print(f"Plaintext: {plaintext}")
print(f"Encrypted: {encrypted_text}")
print(f"Decrypted: {decrypted_text}")
```
