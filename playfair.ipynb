```python
def generate_key_table(key):
    key = "".join(dict.fromkeys(key.replace("j", "i")))
    alphabet = "abcdefghiklmnopqrstuvwxyz"
    key += "".join(c for c in alphabet if c not in key)
    return [list(key[i:i+5]) for i in range(0, 25, 5)]

def format_text(text):
    text = text.replace("j", "i")
    formatted = ""
    i = 0
    while i < len(text):
        formatted += text[i]
        if i + 1 < len(text) and text[i] == text[i + 1]:
            formatted += "x"
        else:
            i += 1
        i += 1
    if len(formatted) % 2:
        formatted += "x"
    return formatted

def find_positions(pair, key_table):
    positions = []
    for char in pair:
        for row in range(5):
            if char in key_table[row]:
                positions.append((row, key_table[row].index(char)))
                break
    return positions

def playfair_encrypt_decrypt(text, key, encrypt=True):
    key_table = generate_key_table(key)
    text = format_text(text)
    result = ""
    shift = 1 if encrypt else -1
    for i in range(0, len(text), 2):
        a, b = text[i], text[i+1]
        pos_a, pos_b = find_positions((a, b), key_table)
        if pos_a[0] == pos_b[0]:
            result += key_table[pos_a[0]][(pos_a[1] + shift) % 5]
            result += key_table[pos_b[0]][(pos_b[1] + shift) % 5]
        elif pos_a[1] == pos_b[1]:
            result += key_table[(pos_a[0] + shift) % 5][pos_a[1]]
            result += key_table[(pos_b[0] + shift) % 5][pos_b[1]]
        else:
            result += key_table[pos_a[0]][pos_b[1]]
            result += key_table[pos_b[0]][pos_a[1]]
    return result

key = "playfairexample"
plaintext = "hide the gold in the tree stump"
formatted_text = plaintext.replace(" ", "").lower()
ciphertext = playfair_encrypt_decrypt(formatted_text, key, encrypt=True)
decrypted_text = playfair_encrypt_decrypt(ciphertext, key, encrypt=False)
print("Ciphertext:", ciphertext)
print("Decrypted Text:", decrypted_text)
```
